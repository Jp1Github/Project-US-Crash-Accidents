#### Converting to the correct data types

The data size originally was at 2+ GB and after dropping it was reduced to 663.3+ MB.
However, the Start_Time and End_Time has the object datatype.

```python
# Single column conversion
# format = ''%Y-%M-%D %H:%M:%S' or 'ISO8601' or 'mixed'
# pd.to_datetime(df['Start_Time'], format='ISO8601')

'''Converting the Start_Time & End_Time column data type from object to datetime64[ns]'''
# Assign a variable to the slices of columns 'Start_Time to 'End_Time'
cols = df.columns[1:3]

# Apply a function doing a loop or iteration
df[cols] = df[cols].apply(pd.to_datetime, errors='coerce')
```

```python
'''Convert the Street, City, County and State datatype from object to Category'''
# Assign a variable to the columns stated above
cols= df.columns[5:9]

# # Apply a function to convert to object to Category
df[cols] = df[cols].apply(pd.Categorical)
```

```python
# Re check the data frame information and size
df.info()
```

During initial view of the data the Severity is just one digit.
Below will check the values in the column to decide which data type is appropriate.

```python
# Check the minimum and maximum values in the Severity column
Severity_min =df['Severity'].min()
Severity_max =df['Severity'].max()

print(f'The Severity min value is {Severity_min} and max values is {Severity_max}')
```

```python
'''
Since the values in the Severity column is just 1 digit.
Decided to convert it to int16. Did not convert to int8 just to give an allowance.
'''
# Convert the Severity data type from int64 to int16
df['Severity'] = df['Severity'].astype('int16')
```

```python
# Re check the data frame info and size
df.info()
```

#### Check null values

```python
# Check the number of counts of empty data on each columns
df.isna().sum()
```

Above output is not easy to read. Decided to make a output
that is easy to read and will compute the ratio of empty data rows of each column.

```python
# Put the output of empty rows into variable col_dict
col = df.isna().sum()

# Convert the output to a dictionary
col_dict = col.to_dict()

# Get the maximum row count and put into a variable
num_of_rows = df.count().max()

# Print out the header/title
print('Missing Values count:')

# Looping into the list
for key, value in col_dict.items():
    # Check if row is not empty
    if value > 0:
        # Check the ratio or percentage agains the total of all row (empty and not empty)
        ratio = (value/num_of_rows)*100
        print(f'{key:<15}\t : {value:,} rows ({ratio:.2f} % of total)')
```

The Street and City columns empty data is just less than 1%. However, the 
Start_Time and End_Time is almost 10%. 

```python
# Time delta between the start and last record date
df['Start_Time'].max() - df['Start_Time'].min()
```

#### Fill in the missing values in Start_Time and End_Time

```python
'''Fill in the values of columns 'Start_Time' and 'End_Time using interpolation'''

# Passing the filled values by interpolation
df[['Start_Time', 'End_Time']] = df[['Start_Time', 'End_Time']].interpolate(method='linear')

# Check if there is no more null values
df.isna().sum()
```

#### Dropping rows of empty data in columns 'Street' and 'City'

```python
# Drop the cells of Street and City column that is NaN
df = df.dropna(subset=['Street', 'City'])

# Alternative using inplace
# df.dropna(subset=['Street', 'City'], inplace=True)
```

```python
# Sanity check that all rows have no more missing values
df.isna().sum()
```

No more missing data in all columns and rows.
